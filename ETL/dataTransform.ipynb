```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("gene.csv")
df.fillna("", inplace=True)
df.head()
```

```python
to_drop = ['Entrez GeneId','Genome Location','Hallmark','Chr Band','Cancer Syndrome',
           'Translocation Partner','Other Germline Mut', 'Other Syndrome', 'Synonyms', 'Tumour Types(Germline)']
df.drop(to_drop,inplace=True, axis=1)
```

```python
df.head()
```

```python
df['Mutation Types'].value_counts()[:5]
```

```python
df['Mutation Types'] = df['Mutation Types'].replace('T', 'Translocation', regex=True)
df['Molecular Genetics'] = df['Molecular Genetics'].replace("", 'NA', regex=True)
df['Somatic'] = df['Somatic'].replace("", 'No', regex=True)
df['Germline'] = df['Germline'].replace("", 'No', regex=True)
```

```python
df.head()
```

```python
df.to_csv("geneData.csv",index=False)
```

```python
somatic_df = df.loc[df['Somatic'] == "yes"]
somatic_df = somatic_df.drop(['Germline', 'Somatic','Tumour Types(Germline)'], axis=1)
somatic_df.head()
```

```python
somatic_df.rename(columns={'Tumour Types(Somatic)': 'Tumour Types'}, inplace=True)
somatic_df.head()
```

```python
somatic_df.to_csv("somatic.csv",index=False)
```
